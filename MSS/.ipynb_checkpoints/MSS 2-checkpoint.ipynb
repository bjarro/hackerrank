```python
def subset_with_highest_sum(numbers_list, target_sum):
    n = len(numbers_list)
    dp = [0] * (target_sum + 1)
    dp[0] = []

    for num in numbers_list:
        for j in range(target_sum, num - 1, -1):
            if dp[j - num] and (not dp[j] or len(dp[j - num]) + 1 > len(dp[j])):
                dp[j] = dp[j - num] + [num]

    return dp[target_sum]

# Example usage:
numbers = [1, 2, 3, 4, 5]
target = 8
result_subset = subset_with_highest_sum(numbers, target)
print("Subset with highest sum:", result_subset)
print("Sum of the subset:", sum(result_subset))
```

```python
def subset_with_least_sum(numbers, target):
    # Step 1: Sort the list in ascending order
    numbers.sort()
    
    global dp_table, min_sum, subset

    # Step 2: Create the dynamic programming table
    dp_table = [float('inf')] * (target + 1)
    dp_table[0] = 0

    # Step 3 and 4: Fill the dynamic programming table
    for num in numbers:
        for i in range(target, num - 1, -1):
#             print(min(dp_table[i], dp_table[i - num] + 1))
            dp_table[i] = min(dp_table[i], dp_table[i - num] + 1)

    # Step 5: Find the smallest sum greater than or equal to the target number
    min_sum = target
    while dp_table[min_sum] == float('inf'):
        print(min_sum)
        min_sum += 1

    # Step 6: Reconstruct the subset
    subset = []
    current_sum = min_sum
    for num in reversed(numbers):
        while current_sum >= num and dp_table[current_sum] == dp_table[current_sum - num] + 1:
            subset.append(num)
            current_sum -= num

    return subset

# Example usage:
# numbers = [3, 1, 9, 4, 7, 2]
numbers = [15, 6, 8, 17, 13, 21]
target = 10
result_subset = subset_with_least_sum(numbers, target)
print(result_subset)  # Output: [3, 7]
```

```python
def subset_with_least_sum(numbers, target):
    # Step 1: Sort the list in ascending order
    numbers.sort()

    # Step 2: Create the dynamic programming table and initialize it
    max_possible_sum = sum(numbers)
    dp_table = [max_possible_sum + 1] * (max_possible_sum + 1)
    dp_table[0] = 0

    # Step 3 and 4: Fill the dynamic programming table
    for num in numbers:
        for i in range(max_possible_sum, num - 1, -1):
            dp_table[i] = min(dp_table[i], dp_table[i - num] + 1)

    # Step 5: Find the smallest sum greater than or equal to the target number
    min_sum = target
    while dp_table[min_sum] == max_possible_sum + 1:
        min_sum += 1

    return subset

# Example usage:
numbers = [15, 6, 8, 17, 13, 21]
target = 10
result_subset = subset_with_least_sum(numbers, target)
print(result_subset)  # Output: [4, 7]
```

```python
def find_subset_with_min_sum(numbers, target):
    n = len(numbers)
    mid = n // 2

    sums_A = []
    for mask in range(1 << mid):
        subset_sum = sum(numbers[i] for i in range(mid) if (mask & (1 << i)) != 0)
        sums_A.append(subset_sum)

    sums_B = []
    for mask in range(1 << (n - mid)):
        subset_sum = sum(numbers[mid + i] for i in range(n - mid) if (mask & (1 << i)) != 0)
        sums_B.append(subset_sum)

    sums_B.sort()

    min_diff = float('inf')
    for sum_A in sums_A:
        idx = bisect_left(sums_B, target - sum_A)
        if idx < len(sums_B):
            min_diff = min(min_diff, sum_A + sums_B[idx] - target)

    return target + min_diff

# Example usage:
numbers = [15, 6, 8, 17, 13, 21]
target = 15
result = find_subset_with_min_sum(numbers, target)
print(result)
```

# Highest sum dp

```python
def subset_with_highest_sum(nums, target):
    n = len(nums)
    dp = [[False] * (target + 1) for _ in range(n + 1)]

    for i in range(n + 1):
        dp[i][0] = True

    for i in range(1, n + 1):
        for j in range(1, target + 1): 
            if nums[i - 1] <= j: # cur_num might sum up to j
                #Sum of previous i or if sum - j is True
                dp[i][j] = dp[i - 1][j] or dp[i - 1][j - nums[i - 1]]
            else: #cur_num cannot sum to j (exceeds j)
                dp[i][j] = dp[i - 1][j]

    highest_sum = 0
    for j in range(target, -1, -1):
        if dp[n][j]:
            highest_sum = j
            break

    subset = []
    i, j = n, highest_sum
    while i > 0 and j > 0:
        if dp[i - 1][j]:
            i -= 1
        else:
            subset.append(nums[i - 1])
            j -= nums[i - 1]
            i -= 1

    return highest_sum, subset

# Example usage:
nums = [1, 3, 4, 6, 8]
target = 10
max_sum, subset = subset_with_highest_sum(nums, target)
print("Subset with the highest sum:", subset)
print("Sum of the subset:", max_sum)
```
