## Func

```python
def maximumSum(a, m):
       
    global list_cumsum, list_cumsum_mod, a_mod
    
    n = len(a)
    
    a_mod = [x % m for x in a]

    
    cumsum = 0
    max_sum = 0
    
    
    list_cumsum = [0]
    list_cumsum = []

    
    for i, num in enumerate(a_mod):
        cumsum += num
        list_cumsum.append(cumsum)
        
        
    list_cumsum_mod = sorted([(c % m, i) for i,c in enumerate(list_cumsum)])
    
    m_end = floor(list_cumsum[-1] / m) + 1
    
    
#     print(list_cumsum)
    
    max_sum = max(list_cumsum_mod, key = lambda tup: tup[0])[0]
    
    
#     print(f'm_end = {m_end}')
#     print(f'max_sum = {max_sum}')
    
#     print(f'-----------------------------------------')
    
    for i, cur_val in enumerate(list_cumsum):
        
#         print(f'Iteration {i}')
        
#         print(list_cumsum_mod)
        
        
        prev_cumsum = list_cumsum[i]
        m_prev_cumsum = prev_cumsum % m
        
        if m_prev_cumsum == 0:
            m_prev_cumsum = m
        
#         Find index where cumsum >= m
#         m_start = ceil(prev_cumsum / m)
        m_start = floor(prev_cumsum / m) + 1
        first_index = bisect_left(list_cumsum, m_start*m)
        
        
#         print(list_cumsum_mod)
        
        i_closest = bisect_left(list_cumsum_mod, m_prev_cumsum-1, key = lambda tup: tup[0])
        
            
#         print(f'm_start = {m_start}')
#         print(f'first_index = {first_index}')
#         print(f'm_prev_cumsum = {m_prev_cumsum}')
#         print(f'searching for [{m_prev_cumsum-1}] to get [{m - 1}]')
        
#         if i_closest >= len(list_cumsum_mod):
        if i_closest >= len(list_cumsum_mod):
#             print(f'index = [{i_closest}]') 
#             print('continue')
            continue
            
            
        #right is closest -> get left (closest left)
        if list_cumsum_mod[i_closest][0] > m_prev_cumsum-1:
            if i_closest >= 1:
                i_closest-=1
            
            else:

#                 print(f'closest_val=[{list_cumsum_mod[i_closest][0]}]')
#                 print(f'm_prev_cumsum-1 = [{m_prev_cumsum-1}]') 
#                 print('i_closest is negative?')
#                 print(f'index = [{i_closest}]') 
#                 print('continue')
                continue
                
        i_cur_invalid = False
            
        i_cur = i_closest
        while list_cumsum_mod[i_cur][1] < first_index:
            if i_cur > 0:
                i_cur -= 1
            else: 
#                 print('no i_cur available - Index 0')
#                 print('breaking - should skip')
                i_cur_invalid = True
                break
                
                
        if list_cumsum_mod[i_cur][0] > m_prev_cumsum-1 or i_cur_invalid:
#             print('no i_cur available')
#             print('continue')
            continue
            
#         if i_cur != i_closest:
#             print(f'i_closest = [{i_closest}]')
#             print(f'i_cur = [{i_cur}]')
            


#         print(f'found [{list_cumsum_mod[i_cur][0]}] got [{list_cumsum_mod[i_cur][0] - m_prev_cumsum + m}]')
#         print(f'index = [{i_cur}]') 
            
        cur_max = list_cumsum_mod[i_cur][0] - m_prev_cumsum + m
        max_sum = max(cur_max, max_sum)
        
        
#         print(f'cur_max = {cur_max}')
#         print(f'max_sum = {max_sum}')
        
#         print(f'-----------------------------------------')
        
    print(max_sum)
        
        
    return max_sum
```

```python
class Suppress:
    def __init__(self, *, suppress_stdout=True, suppress_stderr=False):
        self.suppress_stdout = suppress_stdout
        self.suppress_stderr = suppress_stderr
        self.original_stdout = None
        self.original_stderr = None

    def __enter__(self):
        import sys, os
        devnull = open(os.devnull, "w")

        # Suppress streams
        if self.suppress_stdout:
            self.original_stdout = sys.stdout
            sys.stdout = devnull

        if self.suppress_stderr:
            self.original_stderr = sys.stderr
            sys.stderr = devnull

    def __exit__(self, *args, **kwargs):
        import sys
        # Restore streams
        if self.suppress_stdout:
            sys.stdout = self.original_stdout

        if self.suppress_stderr:
            sys.stderr = self.original_stderr
```
