## Function

```python
def minTime(machines, goal):
    
    print(goal)
    
    global m_sorted, m_cumsum
    
    m_sorted = sorted(machines)
    
    m_cumsum = []
    
    cum_sum = 0
    prev_m = 1
    for m in m_sorted:
        cur_sum = 1 + cum_sum * m/prev_m
        m_cumsum.append(floor(cur_sum))
        cum_sum = cur_sum
        prev_m = m 
        
    print(m_sorted)
    print(m_cumsum)
        

    # ------------
    if goal > m_cumsum[-1]:
        #Use coeff and goal to find initial
        coeff = sum([1/eff for eff in m_sorted])
        initial = floor(goal/coeff)
    #     initial = floor(goal*coeff)
        cur_actual = eval_all(m_sorted, initial)
        cur_total = cur_actual

        print(f'coeff = {coeff}')

        cur_days = initial
        cur_goal = goal - cur_actual # (not keeping track of previous days)
        #cur_goal = goal
    
    else:
        cur_goal = goal
        cur_days = 0
        
    
    print(f'initial days = {initial}')
    print(f'cur_actual = {cur_actual}')
    print(f'cur_total = {cur_total}')
    print(f'cur_goal = {cur_goal}')
    print(f'------------------------')
    
    # ------------
    
    cur_i = bisect_right(m_cumsum, cur_goal)
    i_rightmost = cur_i
    cur_m = m_sorted[cur_i]
    cur_cumsum = m_cumsum[cur_i] # cur_cumsum unused

    adv_days = cur_m - (cur_days % cur_m)
    #adv_days = cur_days % cur_m
    
#     cur_days += adv_days
    try_cur_days = cur_days + adv_days
    
    cur_actual = eval_all(m_sorted, try_cur_days)
    #cur_actual = eval_all(m_sorted, initial)
    
    
    print(f'cumsum > rem = {cur_cumsum}')
    print(f'cur_i = {cur_i}')
    print(f'try_cur_days = {try_cur_days}')
    print(f'cur_actual = {cur_actual}')
    print(f'cur_goal = {cur_goal}')
    print(f'------------------------')
    
    found_greater = False
    i_rightmost = len(m_sorted)
    closest = 0
    
    loop_count = 0 # Temp
    
    while cur_goal > 0 :
        # if actual is less than goal - Use right
        if cur_actual < cur_goal:
            print('in less than')
            cur_i = cur_i+1
            
            if found_greater and cur_i >= i_rightmost:
                print('reached rightmost')
                return floor(closest)
        # if actual is greater than goal
        # Use left and remove machines cur+, 
        #rebase search: #Search lowest cum_sum >= cur_goal   / - actual(cur) 
        
        elif cur_actual > cur_goal:
            print('in greater than')
            found_greater = True
            #Store values
            
            closest = cur_days
            
            # ------------
            #Use left:
            
            print('in use left')
            
            cur_i -= 1
            i_rightmost = cur_i
            cur_m = m_sorted[cur_i]
            cur_cumsum = m_cumsum[cur_i]
            
            #remaining
            adv_days = cur_m - (cur_days % cur_m)
            
            #prev + remaining
            cur_days += adv_days
            
            print(f'cur_days = {cur_days}')
            print(f'cur_m = {cur_m}')
            print(f'adv_days = {adv_days}')
            
            print(f'cur_i = {cur_i}')
            print(f'cur_goal = {cur_goal}')
            
            cur_actual = eval_all(m_sorted, cur_days)  #get next (include leftover days from prev search)
            cur_total = cur_actual
            cur_goal = goal - cur_actual
            
            print(f'cur_actual = {cur_actual}')
            print(f'cur_total = {cur_total}')
            print(f'new_goal = {cur_goal}')
            print(f'------------------------')
            
            # ------------
        
            cur_i = bisect_right(m_cumsum, cur_goal)
            
#             adv_days = cur_days % cur_m
#             if adv_days == 0:
#                 print('in adv_days == 0')
#                 adv_days = cur_m
# #                 cur_i -= 1
# #                 i_rightmost = cur_i
# #                 cur_m = m_sorted[cur_i]
# #                 adv_days = cur_days % cur_m

                
#             print(f'adv days {adv_days}')
#             cur_days += adv_days
            
        i_rightmost = cur_i
        cur_m = m_sorted[cur_i]
            
        adv_days = cur_days % cur_m
        try_cur_days = cur_days + adv_days
#         cur_days += adv_days
#         cur_actual = eval_all(m_sorted, cur_days)
        cur_actual = eval_all(m_sorted, try_cur_days) 
        loop_count+=1
        
        
        print(f'cur_actual {cur_actual}')
        print(f'cur_goal {cur_goal}')
        print(f'try_cur_days {try_cur_days}')
        
        
        if loop_count > 25:
            print('loop count exceeded')
            break
    
    return floor(cur_days)
```
