```python
from collections import Counter
```

```python
#!/bin/python3


# Complete the countTriplets function below.
def countTriplets(arr, r):
    
    dict_counts = Counter(sorted(arr))

    list_counts = dict_counts.items()

    list_progression = []

    dict_lookup = {}

    i = 0

    while dict_counts:

        k = next(iter(dict_counts))
        dict_progression_counts = {}
        dict_progression_counts[k] = dict_counts[k]

        dict_counts.pop(k)
        dict_lookup[k] = i

        while k*r in dict_counts:
            k *= r
            dict_progression_counts[k] = dict_counts[k]
            dict_counts.pop(k)
            dict_lookup[k] = i

        list_progression.append(dict_progression_counts)
        i += 1

    total_count = 0

    for k in arr: 
        i_p = dict_lookup[k]
        k_dict_counts = list_progression[i_p]

        k2 = k*r
        k3 = k2*r

        k_dict_counts[k] -= 1

        if not (k2 in k_dict_counts and k3 in k_dict_counts):
            continue

        c_k2 = k_dict_counts[k2]
        c_k3 = k_dict_counts[k3]

        total_count += c_k2 * c_k3
        
        

    return total_count

    


```

# Step by Step

```python
name_test = 'test_6.txt'

f_input = open(f"./in/{name_test}", "r")

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    nr = f_input.readline().rstrip().split()

    n = int(nr[0])

    r = int(nr[1])

    arr = list(map(int, f_input.readline().rstrip().split()))
```

```python
from collections import OrderedDict
```

```python
%%time
set_k = set(arr)
dict_unique = OrderedDict.fromkeys(sorted(set_k))
```

```python

list_progression = []
dict_lookup = {}
i = 0

while dict_unique:
    
    k = next(iter(dict_unique))
    dict_unique.pop(k)
    dict_lookup[k] = i
    list_sequence = [k]

    while k*r in dict_unique:
        k *= r
        dict_unique.pop(k)
        dict_lookup[k] = i
        list_sequence.append(k)

    list_progression.append(list_sequence)
    i += 1
    
    
list_progression = [sublist for sublist in list_progression if len(sublist) >=3]
set_valid = set([item for sublist in list_progression for item in sublist])
```

```python
%%time

arr_f = [x for x in arr if x in set_valid]

dict_counts_k =  {k:[] for k in set_valid}

for i, k in enumerate(arr_f):
    if dict_counts_k[k]:
        cur_count = dict_counts_k[k][-1][1]
    else:
        cur_count = 0
        
    dict_counts_k[k].append((i, cur_count+1)) 
```

```python
def find_less(i_k1_start, i_k2, list_k1):
    for i in range(i_k1_start, len(list_k1)):
        i_k1, k1 = list_k1[i]
        if i_k1 > i_k2:
            return i - 1
        
    return -2
        
```

```python
dict_counts_twin = {}

for g_p in list_progression:
    for i in range(len(g_p)-1):
        k1 = g_p[i]
        k2 = g_p[i+1]
        dict_counts_twin[k1,k2] = [(0,0)]
        
        i_k1 = 0
        
        list_k1 = dict_counts_k[k1]
        list_k2 = dict_counts_k[k2]
        
        for cur_i, c_k2 in list_k2:

            prev_c = dict_counts_twin[k1,k2][-1][1]
            
            i_k1 = find_less(i_k1, cur_i, list_k1)
            c_k1 = list_k1[i_k1][1]
            
            cur_c = c_k1 * c_k2
            
            dict_counts_twin[k1,k2].append((cur_i, prev_c+cur_c))   
```
