# Evaluator

```python
def deci_eval(decimal):
    len_d = len("%i" % decimal)
    
    value = 0
    
    for i in range(len_d):
        digit = get_digit(decimal, i)
        
        value += digit * 2**i
    return value  
```

```python
def mod_size(i):
    return len("%i" % i)
```

```python
def get_digit(number, n):
    return number // 10**n % 10
```

```python
#How to walk decibinary

def generateN(n):
    list_decibinary = [([0],0),([1],1),([2,0],2)] #(str(decibinary), value)
    
    cur_value = 2
    pv_max = 2
    pv_min = 1
    pv_last = 0
    pv_min_next = 10
    
    i = 4
    heap = []
    
    while i < n:
        
        cur_decibinary = list_decibinary[-1][0].copy()
        
        logger.debug('-------------------------')
        logger.debug(f'iteration [{i}]')
        
        logger.debug(cur_decibinary[pv_last])
        logger.debug(cur_decibinary[pv_last-1])
        
    
        #list_next_pv = [] #(pv_next, decibinary)
        
        #pv_last -> to decrease

        
        if heap:
            logger.debug('in case 1')
            
            cur_decibinary, remainder, pv_last, has_visited_inner = heap[0] #peek
            #pv_last = list_next_pv[0]

            cur_decibinary[pv_last] -=2
            cur_decibinary[pv_last+1] +=1

            
            #Push Inner
            #if pv_last > 0 and  cur_decibinary[pv_last-1] >= 2 and cur_decibinary[pv_last] < 9:
            if not has_visited_inner and pv_last > 0 and cur_decibinary[pv_last-1] >= 2 and cur_decibinary[pv_last] < 9:
                
                heap[0] = (cur_decibinary, remainder, pv_last, True)
                
                pv_last -= 1
                
                cur_decibinary = cur_decibinary.copy()
                cur_decibinary[pv_last] -=2
                cur_decibinary[pv_last+1] +=1
                
                to_push = (cur_decibinary, 0, pv_last, False)
                
                heappush(heap, to_push)
            
            #Compute current then push next (if possible)
            else: 

                #Compute innermost
                pv_next = heappop(heap) 
                list_decibinary.append(pv_next[0])
                i += 1
                
                #Push next
                if  pv_last >= 0 and cur_decibinary[pv_last] >= 2 and cur_decibinary[pv_last+1] < 9:
                    cur_decibinary = cur_decibinary.copy()
                    cur_decibinary[pv_last] -=2
                    cur_decibinary[pv_last+1] +=1
                    
                    to_push = (cur_decibinary, 0, pv_last, has_visited_inner)
                    
                    heappush(to_push)


        else:
            
            cur_value +=1
            
            if cur_value == 2**(pv_max): #cur_value exceeds current pv_max
                pv_max+=1
                

            if cur_value == pv_min_next:
                pv_min +=1
                pv_min_next = pv_min_next + 9*2**pv_min

            cur_decibinary = [0 for _ in range(pv_max)]
            
            #-------  Build
            if cur_value % 2 == 0:
                first_digit = min(cur_value,8)
                
            else:
                first_digit = min(cur_value,9)
            
            cur_decibinary[0] = first_digit
            

            rem_value = cur_value - first_digit
            
            for j in range(1, pv_min):

                cur_rem = rem_value // 2**j
                cur_digit = min(cur_rem,9)
                
                if cur_digit:
                    cur_decibinary[j] = cur_digit
    
                rem_value = rem_value - cur_rem*2**j
       
            #Push all levels (same cur_decibinary, different pv)
            
            for j in range(pv_max):
                to_push = (cur_decibinary, 0, j, False)
                heappush(heap, to_push)
                
                
                
    return list_decibinary


```
