```python
def generateN(n):
    
    global ds, ds2

    ds = defaultdict(list)
    ds.update({0:[(0,1)], 1:[(1,2)], 2:[(2,3),(10,4)]})
    
    #dict of dict or dict of list?
    ds2 = defaultdict(lambda: defaultdict(list))
    ds2.update({0:{0:[0]}, 1:{0:[1]}, 2:{0:[2,10], 1:[10]}})
    
    #ds2.update({0:[], 1:[], 2:[[2,10], [10]]})
    
    i = 4
    n_pv = -1
    cur_value = 2
    next_max = 4
    cur_pv = 2

    
    while i < n:
        
        cur_value += 1
        cur_list = ds[cur_value]
        
#         if cur_value == 50:
#             break
        
        logger.debug(f'------------------------') 
        logger.debug(f'in next value {cur_value}')           
        #Next PV
        
        if cur_value == next_max:
            is_even = True

            logger.debug(f'---- increase pv') 
            cur_pv += 1

            n_pv = cur_pv-2
            logger.debug(f'n_pv = {n_pv}') 

#             if i == n:
#                 break

        elif cur_value % 2 == 0:

            is_even = True

            ##Factor cur_value

            list_bits = [(cur_value >> bit) & 1 for bit in range(cur_pv - 1, -1, -1)]

            logger.debug(f'in factor cur_value {list_bits}') 
            
            #Convert cur_val to binary
            #Find pv of first occurence of 1
            for j, bit in enumerate(reversed(list_bits)):
                if bit == 1:
                    n_pv = j
                    logger.debug(f'n_pv = {n_pv}') 
                    break

        else:
            is_even = False
            n_pv = 0


        logger.debug(f'in i = {i}') 
        
        #Initialize to_add
        for j in range(n_pv+1):
            
            logger.debug(f'------') 
            
            logger.debug(f'in j = {j}') 
                
            
            if j == 0:
                if is_even:
                    to_add = 2
                    key_to_increase = cur_value - 2
                
                if not is_even:
                    to_add = 1
                    key_to_increase = cur_value - 1
                    
            else:
                to_add = 10**j
                key_to_increase = cur_value - 2**(j)
                        
            
            logger.debug(f'in key = {key_to_increase}, {j}') 

                
            list_to_increase = ds2[key_to_increase][j]

            for to_increase in list_to_increase:
                         
                res = to_increase + to_add
                
                #get digit
                jth_digit_s = to_increase // 10**(j+1) % 10
                jth_digit_e = res // 10**(j+1) % 10
                
                if jth_digit_s != jth_digit_e:
                    continue
                
                i+=1
                
                ds[cur_value].append((res,i))
                
                for k in range(j+1):
                     ds2[cur_value][k].append(res)
                    
                
                #ds2[cur_value][j].append(res)
                         
                logger.debug(f'[{(res,i)}] = [{to_increase}] + [{to_add}]') 
                
                
                #ds2[cur_value][j] = sorted(ds2[cur_value][j])
                
                if i == n:
                    break
                    
            if i == n:
                break 
                
        if cur_value == next_max:
            logger.debug(f'---- prev_unique x2')
            i += 1
            to_append = (10**(cur_pv-1), i)
            cur_list.append(to_append) 

            for k in range(j+2):
                ds2[cur_value][k].append(to_append[0])

            next_max *= 2


        for k in range(j+1):
             ds2[cur_value][k] = sorted(ds2[cur_value][k])


        ds[cur_value] = sorted(ds[cur_value])
                

        if i == n:
            break
            

    return ds
```
